```python
import matplotlib.pyplot as plt
import numpy as np

fig,axes = plt.subplots(3,4,figsize = (40,30))
a = np.linspace(1,6,10)
for i in range(3):
    axes[0,i].plot(a, a)
    axes[1,i].plot(a,a)
    axes[1,i].plot(a,a)

    axes[0,i].set_title("Title")

plt.setp(axes[0,0],ylabel="Distribution ")
plt.setp(axes[1,0],ylabel="Acceptance ")
plt.setp(axes[2,0], ylabel = "Convergence")
    
plt.show()
```
